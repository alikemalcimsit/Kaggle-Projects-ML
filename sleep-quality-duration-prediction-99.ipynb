```python
df=pd.read_csv("/kaggle/input/sleep-health-and-digital-screen-exposure-dataset/Dry_Eye_Dataset.csv")
df.head(3)
```

```python
df.columns.tolist()
```

```python
df=df.drop([  'Blue-light filter',
 'Discomfort Eye-strain',
 'Redness in eye',
 'Itchiness/Irritation in eye',
 'Dry Eye Disease'],axis=1)
df.head(3)
```

```python
df.dtypes
```

```python

columns_to_replace = [
    'Sleep disorder', 
    'Wake up during night', 
    'Feel sleepy during day', 
    'Caffeine consumption', 
    'Alcohol consumption', 
    'Smoking', 
    'Medical issue', 
    'Ongoing medication', 
    'Smart device before bed'
]

df[columns_to_replace] = df[columns_to_replace].replace({'Y': 1, 'N': 0}).astype(int)

# SonuÃ§larÄ± kontrol et
df.head(3)
```

```python
df.dtypes
```

```python
# 'Blood pressure' sÃ¼tununu '/' karakterine gÃ¶re ayÄ±r ve iki yeni kolona ata
df[['Blood pressure systolic', 'Blood pressure diastolic']] = df['Blood pressure'].str.split('/', expand=True).astype(int)
df=df.drop(["Blood pressure"] , axis=1)
# SonuÃ§larÄ± kontrol et
df.head(3)
```

```python
df.dtypes
```

```python
df= pd.get_dummies(df,["Gender"], drop_first=True)
df.head(3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Sleep Quality Prediction
x=df.drop(["Sleep quality"],axis=1)
y=df["Sleep quality"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
print(model.predict(x_test))
```

```python
#Sleep Duration Prediction
x=df.drop(["Sleep duration"],axis=1)
y=df["Sleep duration"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
print(model.score(x_test,y_test))
print(model.predict(x_test))
x_test.head(1)
```
